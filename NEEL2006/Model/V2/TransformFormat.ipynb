```python
with open('twitter_test.txt') as f:
    lines = f.readlines()
    
word_dict={}

for line in lines:

    line_li=line.split('|')
    id=line_li[1]
    word=line_li[3]
    label=line_li[4]

    if word not in word_dict:
        word_dict[word] = label
```

```python
with open('final_test_sentence.txt') as t:
    sentences = t.readlines()

dedup_sentences=[]

tmp=''
for sen in sentences:
    clear_sen = sen.strip()
    if clear_sen != tmp:

        tmp=clear_sen
        dedup_sentences.append(clear_sen)
```

```python
filename = 'test.txt'
with open(filename,'w') as twitter:
    # twitter.write("I am Meringue.\n")

    for sen in dedup_sentences:

        words_sen = sen.split(' ')
        #print(sen)
        for w in words_sen:

            if w in word_dict:
                print(w,word_dict[w])
                twitter.write(w+' '+word_dict[w]+'\n')
            else:
                twitter.write(w + ' ' + '0' + '\n')

        twitter.write('. 0' + '\n')
        #twitter.write( '\n')
```
